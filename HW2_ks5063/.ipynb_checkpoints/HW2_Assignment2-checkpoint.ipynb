 <h2><SPAN STYLE="COLOR:BLUE">2) DATE-TIME AND NUMERIC COLUMNS</SPAN></h2>

```python
#!curl https://data.cityofnewyork.us/api/views/bbs3-q5us/rows.csv?accessType=DOWNLOAD > campaign_contributions.csv
```

```python
#!mv campaign_contributions.csv $PUIDATA
#!ls $PUIDATA
```

```python
campaign_data = pd.read_csv("%s/campaign_contributions.csv"%os.environ["PUIDATA"],low_memory=False)
campaign_data.head(7)
```

```python
campaign_2col = campaign_data.iloc[:,[11,37]]
campaign_2col.head()
```

```python
campaign_2col['DATE'] = pd.to_datetime(campaign_2col.DATE)
campaign_2col = campaign_2col[campaign_2col['AMNT']<10000] #remove outlier
campaign_2col = campaign_2col[campaign_2col['AMNT']>0] #remove negative values
campaign_2col['YEAR'] = campaign_2col['DATE'].map(lambda x: str(x).split('-')[0]) #extract year from date
campaign_2col = campaign_2col[campaign_2col['YEAR']!="NaT"] #remove missing values
campaign_2col['YEAR'] = campaign_2col['YEAR'].map(lambda x: int(x)) #convert YEAR to int
campaign_2col = campaign_2col[(campaign_2col['YEAR']>=2007)&(campaign_2col['YEAR']<=2008)] #subset from YEAR 2008 to 2009
campaign_2col = campaign_2col.sort_values(['DATE'], ascending=[1]) #sort according to DATE
```

```python
ax = campaign_2col.plot(x='DATE', y='AMNT',rot=90)
ax.set_ylabel("Contribution Amount");
```

FIGURE 2 : Contribution Amount vs DATE as shown by Campaign Contributions 2009 data. There seems to be a periodic trend of donations in a square wave like  pattern. The donations also seem to show a decrease towards the last 2 months of the campaign.

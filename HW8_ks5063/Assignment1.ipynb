```python
import pandas as pd
import geopandas as gpd
import matplotlib.pylab as plt
from matplotlib import pyplot
import shapely
import fiona
import pyproj
import re
import os

try:
    from urllib import urlretrieve as urlretrieve
except ImportError:
    from urllib.request import urlretrieve as urlretrieve

pd.set_option('display.max_columns', 500)
%matplotlib inline
```

```python
df1 = pd.read_csv("https://data.cityofnewyork.us/api/views/jr24-e7cr/rows.csv?accessType=DOWNLOAD") #electric charges
```

```python
df1.head()
```

```python
df2 = pd.read_csv("https://data.cityofnewyork.us/api/views/66be-66yr/rows.csv?accessType=DOWNLOAD") #water charges
```

```python
df2.head()
```

```python
df3 = pd.read_csv("https://data.cityofnewyork.us/api/views/it56-eyq4/rows.csv?accessType=DOWNLOAD") #heat
```

```python
df3.head()
```

```python
df4 = pd.read_csv("https://data.cityofnewyork.us/api/views/avhb-5jhc/rows.csv?accessType=DOWNLOAD") #gas
```

```python
df4.head()
```

```python
df1 = df1.loc[:,["Borough","Current Charges"]]
df2 = df2.loc[:,["Borough","Current Charges"]]
df3 = df3.loc[:,["Borough","Current Charges"]]
df4 = df4.loc[:,["Borough","Current Charges"]]

df1["Current Charges"] = df1["Current Charges"].map(lambda x : re.sub('[$,()]', '', str(x)))
df1["Current Charges"] = df1["Current Charges"].astype(float)
df1.head()
```

```python
df2["Current Charges"] = df2["Current Charges"].map(lambda x : re.sub('[$,()]', '', str(x)))
df2["Current Charges"] = df2["Current Charges"].astype(float)
df2.head()
```

```python
df3["Current Charges"] = df3["Current Charges"].map(lambda x : re.sub('[$,()]', '', str(x)))
df3["Current Charges"] = df3["Current Charges"].astype(float)
df3.head()
```

```python
df4["Current Charges"] = df4["Current Charges"].map(lambda x : re.sub('[$,()]', '', str(x)))
df4["Current Charges"] = df4["Current Charges"].astype(float)
df4.head()
```

```python
df1_agg = df1.groupby('Borough').mean().reset_index()
df1_agg = df1_agg[df1_agg["Borough"].isin(["BROOKLYN","MANHATTAN","BRONX"])]
df1_agg.rename(columns={"Current Charges":"Electric"},inplace=True)
df1_agg["Borough"] = df1_agg["Borough"].map({"BROOKLYN":"Brooklyn","MANHATTAN":"Manhattan","BRONX":"Bronx"})
df1_agg.head()
```

```python
df2_agg = df2.groupby('Borough').mean().reset_index()
df2_agg = df2_agg[df2_agg["Borough"].isin(["BROOKLYN","MANHATTAN","BRONX"])]
df2_agg.rename(columns={"Current Charges":"Water"},inplace=True)
df2_agg["Borough"] = df2_agg["Borough"].map({"BROOKLYN":"Brooklyn","MANHATTAN":"Manhattan","BRONX":"Bronx"})
df2_agg.head()
```

```python
df3_agg = df3.groupby('Borough').mean().reset_index()
df3_agg = df3_agg[df3_agg["Borough"].isin(["BROOKLYN","MANHATTAN","BRONX"])]
df3_agg.rename(columns={"Current Charges":"Heat"},inplace=True)
df3_agg["Borough"] = df3_agg["Borough"].map({"BROOKLYN":"Brooklyn","MANHATTAN":"Manhattan","BRONX":"Bronx"})
df3_agg.head()
```

```python
df4_agg = df4.groupby('Borough').mean().reset_index()
df4_agg = df4_agg[df4_agg["Borough"].isin(["BROOKLYN","MANHATTAN","BRONX"])]
df4_agg.rename(columns={"Current Charges":"Gas"},inplace=True)
df4_agg["Borough"] = df4_agg["Borough"].map({"BROOKLYN":"Brooklyn","MANHATTAN":"Manhattan","BRONX":"Bronx"})
df4_agg.head()
```

```python
df_all = df1_agg.merge(df2_agg,on="Borough")
df_all = df_all.merge(df3_agg,on="Borough")
df_all = df_all.merge(df4_agg,on="Borough")
df_all.head()
```

```python
nyc_shp = os.environ["PUIDATA"] + "/geo_export_8c6ca477-c6f7-482f-85ae-6dba24fb0dca.shp"
nyc_shp = gpd.GeoDataFrame.from_file(nyc_shp)
```

```python
nyc_shp.crs
```

```python
nyc_shp.head()
```

```python
nyc_shp = nyc_shp.merge(df_all,left_on="boro_name",right_on="Borough",how="right")
nyc_shp.head()
```

```python
nyc_shp['boro_code'] = nyc_shp["boro_name"].map({"Manhattan":"MN","Brooklyn":"BK","Bronx":"BR"})
```

```python
fig, axes = pyplot.subplots(ncols=2, nrows=2, figsize=(12,12), sharex=True, sharey=True)
ax1 = nyc_shp.plot(ax=axes[0, 0], column='Electric',cmap="OrRd",edgecolor="black",scheme="Quantiles",)
nyc_shp.apply(lambda state: 
                  ax1.annotate(s=state['boro_code'], 
                              xy=state.geometry.centroid.coords[0], 
                              ha='center', clip_on=False,color="blue"), axis=1)
axes[0, 0].set_title('Average Electric Charges')
ax2 = nyc_shp.plot(ax=axes[0, 1], column='Water',cmap="OrRd",edgecolor="black",scheme="Quantiles")
nyc_shp.apply(lambda state: 
                  ax2.annotate(s=state['boro_code'], 
                              xy=state.geometry.centroid.coords[0], 
                              ha='center', clip_on=False,color="blue"), axis=1)
axes[0, 1].set_title('Average Water Charges')
ax3 = nyc_shp.plot(ax=axes[1, 0], column='Heat',cmap="OrRd",edgecolor="black",scheme="Quantiles")
nyc_shp.apply(lambda state: 
                  ax3.annotate(s=state['boro_code'], 
                              xy=state.geometry.centroid.coords[0], 
                              ha='center', clip_on=False,color="blue"), axis=1)
axes[1, 0].set_title('Average Heat Charges')
ax4 = nyc_shp.plot(ax=axes[1, 1], column='Water',cmap="OrRd",edgecolor="black",scheme="Quantiles")
nyc_shp.apply(lambda state: 
                  ax4.annotate(s=state['boro_code'], 
                              xy=state.geometry.centroid.coords[0], 
                              ha='center', clip_on=False,color="blue"), axis=1)
axes[1, 1].set_title('Average Gas Charges')
fig.suptitle("Average Utility Charges across Brooklyn, Manhattan and Bronx",fontsize=15)
axes[0, 0].set_axis_off()
axes[0, 1].set_axis_off()
axes[1, 0].set_axis_off()
axes[1, 1].set_axis_off()
plt.show()
```

<strong>Figure 1 : Average Utility Charges across Brooklyn, Manhattan and Bronx. Brooklyn seems to have the least utility charges. Manhattan and Bronx are alternatively highest among different utilities.</strong>
